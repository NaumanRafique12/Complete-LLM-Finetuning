```python
import fitz  # pip install pymupdf

def extract_text_from_pdf(pdf_path):
    text_blocks = []
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text = page.get_text("text").strip()
            if text:
                text_blocks.append(text)
    return text_blocks
```

```python
pdf_texts = extract_text_from_pdf("/content/Metformin.pdf")
```

```python
import re

def split_paragraphs(pages):
    paragraphs = []
    for page_text in pages:
        # Split on double line breaks or long newlines
        chunks = re.split(r'\n\s*\n', page_text)
        for chunk in chunks:
            clean = chunk.strip()
            if len(clean) > 30:  # ignore too short lines
                paragraphs.append(clean)
    return paragraphs

paragraphs = split_paragraphs(pdf_texts)
```

```python
data = [{"text": p} for p in paragraphs]
print(data)
```

[ Embedding Layer ]
   ↓
[ Transformer Block 1 ]
[ Transformer Block 2 ]
[ Transformer Block 3 ]
      ...
[ Transformer Block N ]
   ↓
[ LM Head / Output Layer ]


| Approach          | Layers Trained        | Params Trained | GPU Need  | Notes               |
| ----------------- | --------------------- | -------------- | --------- | ------------------- |
| Full Fine-Tune    | All                   | 100%           | 🔥 High   | Best quality, heavy |
| Partial Fine-Tune | Subset (e.g. top 4)   | ~10–20%        | ⚙️ Medium | Balanced trade-off  |
| LoRA / Adapters   | Inserted small layers | <1%            | 🧩 Low    | Most efficient      |


| Method               | Gradient Memory | Compute Cost | Adapter Save Size              | Mergeable? |
| -------------------- | --------------- | ------------ | ------------------------------ | ---------- |
| Freeze Bottom Layers | Moderate        | Medium       | Large (full layer checkpoints) | ❌ Hard     |
| LoRA / QLoRA         | Very Low        | Low          | Tiny (few MBs)                 | ✅ Easy     |
| Full Fine-Tune       | High            | High         | Huge (GBs)                     | ✅ Yes      |


| Point                   | Freeze Layers | LoRA                       |
| ----------------------- | ------------- | -------------------------- |
| Compute saving          | ✅ Medium      | ✅✅✅ Massive                |
| Representation learning | ❌ Limited     | ✅ Deep (within all layers) |
| Merge multiple domains  | ❌ Hard        | ✅ Easy                     |
| Quantization support    | ⚠️ Difficult  | ✅ Works with QLoRA         |
| Adapter modularity      | ❌ None        | ✅ Plug-and-play            |
| Industry adoption       | ⚙️ Low        | 🚀 Standard                |


Layer-freezing ek old-school transfer learning trick hai (used in BERT/CNN era).
LoRA is modern PEFT —
it preserves base knowledge, adapts efficiently, and scales modularly.

```python
from transformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained("TinyLlama/TinyLlama-1.1B-intermediate-step-1431k-3T")

# Freeze all layers
for param in model.parameters():
    param.requires_grad = False

# Unfreeze last 4 transformer blocks and lm_head
for name, param in model.named_parameters():
    if any(f"layers.{i}." in name for i in range(20, 24)):  # last 4 layers
        param.requires_grad = True
    if "lm_head" in name:
        param.requires_grad = True

# Check trainable params
trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
total = sum(p.numel() for p in model.parameters())
print(f"Trainable: {trainable/total*100:.2f}% parameters")
```

```python
from trl import DPOTrainer, DPOConfig
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset

# Model and tokenizer
model_name = "NousResearch/Llama-2-7b-chat-hf"
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load dataset
dataset = load_dataset("json", data_files="your_dpo_dataset.json")["train"]

# Training args
dpo_args = DPOConfig(
    output_dir="./llama2-dpo-finetuned",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    learning_rate=5e-6,
    num_train_epochs=2,
    beta=0.1,
    logging_steps=50,
    save_strategy="epoch"
)

# Trainer
trainer = DPOTrainer(
    model=model,
    ref_model=None,  # or a frozen reference model
    args=dpo_args,
    train_dataset=dataset,
    tokenizer=tokenizer,
)
trainer.train()
```
